```python
import numpy as np
import matplotlib.pyplot as plt
gs = a.groupby('category')['actual_price', 'discounted_price'].sum()
labels = ['actual_price', 'discounted_price']
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += [angles[0]]
categories = gs.index
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))
for i, category in enumerate(categories):
    values = gs.loc[category].values.tolist()
    values += values[:1]
    ax.plot(angles, values, color='blue', linewidth=1)
    ax.fill(angles, values, color='blue', alpha=0.25)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
plt
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
df = pd.read_csv('amazon.csv')
df
```

```python
fig = px.line_polar(
    df,
    theta="rating", 
    r="discounted_price",
    line_close=True,
    range_r=[0, df["actual_price"].max()], 
)
fig.update_layout(title="Product Performance Comparison")
fig.show()
```
